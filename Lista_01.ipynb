# 8. Crie um novo modelo de regressão linear para a feature que melhor se ajusta a uma função linear (ver questão 2a).

```python
# Criação do modelo
model_3 = linear_model.LinearRegression()
x_3 = [[i] for i in X[0:, 3]]
y_3 = [[i] for i in y]
model_3.fit(x_3, y)
```

```python
# Calculo do coeficiente B1
b1 = np.sum((x_3 - np.mean(x_3)) * (y_3 - np.mean(y_3))) / np.sum((x_3 - np.mean(x_3)) ** 2)
print(b1)
```

```python
# Calculo do coeficiente B0
b0 = np.mean(y_3) - b1 * np.mean(x_3)
print(b0)
```

```python
model_3.coef_
```

```python
model_3.intercept_
```

```python
# Grafico
y_pred_3 = model_3.predict(x_3)
plt.grid()
plt.scatter(x_3, y)
plt.plot(x_3, y_pred_3)
```

```python
# Calculo do MSE
mse = metrics.mean_squared_error(y_3, y_pred_3)
print(mse)
```

# 9. Crie um novo modelo de regressão linear para a feature que pior se ajusta a uma função linear (ver questão 2b).

```python
# Criação do modelo
model_1 = linear_model.LinearRegression()
x_1 = [[i] for i in X[0:, 1]]
y_1 = [[i] for i in y]
model_1.fit(x_1, y)
```

```python
# Calculo do coeficiente B1
b_1 = np.sum((x_1 - np.mean(x_1)) * (y_1 - np.mean(y_1))) / np.sum((x_1 - np.mean(x_1)) ** 2)
print(b_1)
```

```python
# Calculo do coeficiente B0
b0 = np.mean(y_1) - b_1 * np.mean(x_1)
print(b0)
```

```python
model_1.coef_
```

```python
model_1.intercept_
```

```python
# Grafico
y_pred_1 = model_1.predict(x_1)
plt.grid()
plt.scatter(x_1, y)
plt.plot(x_1, y_pred_1)
```

```python
# Calculo do MSE
mse = metrics.mean_squared_error(y_1, y_pred_1)
print(mse)
```
